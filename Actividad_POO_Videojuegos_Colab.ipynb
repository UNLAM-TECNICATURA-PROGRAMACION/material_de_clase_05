# Parte 2 — Clase y objeto

Una **clase** es como un molde, plano o receta.

Un **objeto** es algo concreto creado a partir de ese molde.

Por ejemplo:

- Clase: `Personaje`
- Objetos: `Luna`, `Kai`, `Nina`

Todos son personajes, pero cada uno puede tener valores distintos.

## Analogía

Una clase es como un molde para hacer galletitas.

Cada galletita creada con ese molde es un objeto.

Todas tienen una forma parecida, pero pueden tener distinto color, sabor o decoración.

# Parte 3 — Primera clase en Python

Vamos a crear una clase llamada `Personaje`.

Esta clase servirá como molde para crear personajes de un videojuego.

Cada personaje tendrá:

- `nombre`
- `vida`
- `energia`

Y podrá realizar una acción:

- `mostrar_estado()`

```python
class Personaje:
    def __init__(self, nombre, vida, energia):
        self.nombre = nombre
        self.vida = vida
        self.energia = energia

    def mostrar_estado(self):
        print("Nombre:", self.nombre)
        print("Vida:", self.vida)
        print("Energía:", self.energia)
```

## Observemos el código

En el ejemplo anterior:

- `class Personaje` define una clase.
- `__init__` es un método especial que se ejecuta cuando se crea un objeto.
- `self.nombre`, `self.vida` y `self.energia` son atributos.
- `mostrar_estado()` es un método.

Ahora vamos a crear un objeto concreto a partir de la clase `Personaje`.

```python
personaje_1 = Personaje("Luna", 100, 50)

personaje_1.mostrar_estado()
```

# Parte 5 — El estado de un objeto

El **estado** de un objeto es el conjunto de valores que tienen sus atributos en un momento determinado.

Por ejemplo, un personaje puede empezar con 100 puntos de vida.

Si recibe daño, sigue siendo el mismo objeto, pero su estado cambia.

Vamos a agregar dos acciones al personaje:

- `recibir_danio(cantidad)`
- `curarse(cantidad)`

```python
class Personaje:
    def __init__(self, nombre, vida, energia):
        self.nombre = nombre
        self.vida = vida
        self.energia = energia

    def mostrar_estado(self):
        print("Nombre:", self.nombre)
        print("Vida:", self.vida)
        print("Energía:", self.energia)

    def recibir_danio(self, cantidad):
        self.vida = self.vida - cantidad
        print(self.nombre, "recibió", cantidad, "puntos de daño.")

    def curarse(self, cantidad):
        self.vida = self.vida + cantidad
        print(self.nombre, "recuperó", cantidad, "puntos de vida.")
```

```python
heroe = Personaje("Luna", 100, 50)

print("Estado inicial:")
heroe.mostrar_estado()

print("------")
heroe.recibir_danio(20)
heroe.mostrar_estado()

print("------")
heroe.curarse(10)
heroe.mostrar_estado()
```

# Parte 6 — Agregamos reglas al objeto

En un videojuego, no tendría mucho sentido que un personaje tenga vida negativa.

Por ejemplo, si tiene 30 de vida y recibe 100 de daño, no queremos que quede con `-70` de vida.

Vamos a mejorar la clase para que la vida nunca baje de 0.

También agregaremos un método llamado `esta_vivo()`.

```python
class Personaje:
    def __init__(self, nombre, vida, energia):
        self.nombre = nombre
        self.vida = vida
        self.energia = energia

    def mostrar_estado(self):
        print("Nombre:", self.nombre)
        print("Vida:", self.vida)
        print("Energía:", self.energia)

    def recibir_danio(self, cantidad):
        self.vida = self.vida - cantidad

        if self.vida < 0:
            self.vida = 0

        print(self.nombre, "recibió", cantidad, "puntos de daño.")

    def curarse(self, cantidad):
        self.vida = self.vida + cantidad
        print(self.nombre, "recuperó", cantidad, "puntos de vida.")

    def esta_vivo(self):
        if self.vida > 0:
            return True
        else:
            return False
```

```python
heroe = Personaje("Luna", 100, 50)

heroe.recibir_danio(150)
heroe.mostrar_estado()

if heroe.esta_vivo():
    print("El personaje sigue vivo.")
else:
    print("El personaje fue derrotado.")
```

# Parte 7 — Creamos una clase Enemigo

Ahora vamos a crear otra clase: `Enemigo`.

Un enemigo tendrá:

- `nombre`
- `vida`
- `danio`

Y podrá realizar estas acciones:

- `mostrar_estado()`
- `recibir_danio(cantidad)`
- `atacar(personaje)`

La idea importante es que un objeto pueda interactuar con otro objeto.

En este caso, el enemigo podrá atacar al personaje.

```python
class Enemigo:
    def __init__(self, nombre, vida, danio):
        self.nombre = nombre
        self.vida = vida
        self.danio = danio

    def mostrar_estado(self):
        print("Nombre:", self.nombre)
        print("Vida:", self.vida)
        print("Daño:", self.danio)

    def recibir_danio(self, cantidad):
        self.vida = self.vida - cantidad

        if self.vida < 0:
            self.vida = 0

        print(self.nombre, "recibió", cantidad, "puntos de daño.")

    def atacar(self, personaje):
        print(self.nombre, "ataca a", personaje.nombre)
        personaje.recibir_danio(self.danio)
```

```python
heroe = Personaje("Luna", 100, 50)
slime = Enemigo("Slime verde", 40, 15)

print("Estado inicial del héroe:")
heroe.mostrar_estado()

print("------")
print("Estado inicial del enemigo:")
slime.mostrar_estado()

print("------")
slime.atacar(heroe)

print("------")
print("Estado del héroe después del ataque:")
heroe.mostrar_estado()
```
